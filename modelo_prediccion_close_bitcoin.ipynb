```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01,random_state=72)
```

```python
regBDT = DecisionTreeRegressor(max_depth=6,splitter="random")
regBDT.fit(X_train, y_train)

y_pred = regBDT.predict(X_test)

r2_score(y_test, y_pred)
```

```python
from sklearn.neighbors import KNeighborsRegressor

regKNN=KNeighborsRegressor(n_neighbors=30,weights="distance")

regKNN.fit(X_train, y_train)

y_pred1 = regKNN.predict(X_test)

r2_score(y_test, y_pred1)
```

```python
from sklearn.ensemble import RandomForestRegressor

regRFR=RandomForestRegressor(n_estimators=200,max_depth=15)

regRFR.fit(X_train, y_train)

y_pred2 = regRFR.predict(X_test)

r2_score(y_test, y_pred2)
```
